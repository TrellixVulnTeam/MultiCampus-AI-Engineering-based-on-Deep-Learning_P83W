### 2. 다나와 -> 노트북페이지

```python
# - 가격 데이터 추출 및 저장
# - 화면 크기 추출 및 저장
# - 인텔/ AMD... 추출 및 저장
# - 옥타/ 쿼드/ 헥사코어... 추출 및 저장
# - SSD 용량
# - DDR4 추출 및 저장
# - 제조사 추출 및 저장
# => 데이터 프레임으로 저장
```

```python
driver = webdriver.Chrome("C://mychrome/chromedriver.exe")
url = "http://prod.danawa.com/list/?cate=112758&15main_11_02"
driver.get(url)
html = driver.page_source
soup = BeautifulSoup(html, "lxml")
```

```python
notebooks = soup.select("div.prod_main_info")
nb_list = []

for nb in notebooks:
    dic = {}
    nb_name = nb.select("div.prod_info > p.prod_name > a")[0].text.strip()
    dic['Name'] = nb_name
    dic['Manufacturer'] = nb_name.split()[0]
    price = nb.select("div.prod_pricelist > ul >li> p.price_sect > a > strong")[0].text
    dic['Price'] = price
    
    if nb.select("div.prod_info > p.prod_name > strong"):      
        specs = nb.select("div.prod_info > dl.prod_spec_set > dd > div.spec_list > a")     
    else:
        specs = nb.select("div.prod_info > dl.prod_spec_set > dd > ul > li.spec_item > a")

    for i, spec in enumerate(specs):
            if i == 0:
                dic['Size'] = spec.text
            elif i == 1:
                dic['CPU'] = spec.text
            elif re.match("[가-힣0-9]+코어$",spec.text):
                dic['Core'] = spec.text
            elif spec.text == "SSD" and re.match("^[0-9]+[GB|TB]", specs[i+1].text):
                dic['SSD'] = specs[i+1].text
            elif re.search("DDR4", spec.text):
                dic['DDR'] = spec.text
    nb_list.append(dic)
    
nb_list   
```

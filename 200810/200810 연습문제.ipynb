# 1. 인스타그램에서 '제주도맛집' 키워드에 해당되는 게시글 50개를 추출하여 results에 저장

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import re
import pandas as pd
```

```python
def insta_searching(word):
    url = "https://www.instagram.com/explore/tags/"+word
    return url

def select_first(driver):
    first = driver.find_element_by_css_selector("._9AhH0")
    first.click()
    time.sleep(2)
    
def get_content(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, "lxml")
    
    # 게시글 읽기
    try:
        content = soup.select("div.C4VMK > span")[0].text
    except:
        content = "" 

    ## content에서 해시태그만 가져오기
    hashtag =re.findall("#[\w,가-힣ㄱ-ㅎ]+", content) 
    
    #날짜 가져오기
    date = soup.select("time._1o9PC.Nzb55")[0]['datetime'][:10]
    
    #좋아요 수 가져오기
    try:
        if soup.select("div.Nm9Fw > button > span"):
            like = soup.select("div.Nm9Fw > button > span")[0].text
        elif soup.select("div.Nm9Fw > button")[0].text.split()[1] == '1개':
            like = '1'
    except:
        like = ""
    #위치 정보 가져오기
    try:
        place = soup.select("div.M30cS")[0].text
    except:
        place = ""
    data = [content, date, like, place, hashtag]
    return data

def moveNext(driver):
    right = driver.find_element_by_css_selector("a.coreSpriteRightPaginationArrow")
    right.click()
    time.sleep(2)
```

```python
driver = webdriver.Chrome("C://mychrome/chromedriver.exe")
driver.get("https://www.instagram.com")
```

```python
word = "제주도맛집"
url = insta_searching(word)
driver.get(url) #한글을 유니코드 형식으로 읽어주는 함수 : get
time.sleep(2)
```

```python
select_first(driver)
```

```python
results = []
for i in range(50):
    data = get_content(driver)
    results.append(data)
    moveNext(driver)
```

```python
pd.DataFrame(results, columns=["Content", "date", "like", "place", "tags"])
```

# 2. 음수가 아닌 수들이 주어졌을 때 그 수들을 이어서 만들 수 있는 가장 큰 수를 구하시오. 
예를 들어 [1,2,3]이 주어졌을 때 만들 수 있는 가장 큰 수는 321이고, [3, 30, 34, 5, 9] 가 주어지면 만들 수 있는 가장 큰 수는 9534330이다.

```python
import operator

num_list= list(map(str, input("수를 입력하세요 : ").split()))

dic = {}
for i, num in enumerate(num_list):
    dic[i] = num*2
dic = sorted(dic.items(), key=operator.itemgetter(1), reverse=True)

answer = ""
for i in range(len(dic)):
    answer += num_list[dic[i][0]]
print(answer)

```

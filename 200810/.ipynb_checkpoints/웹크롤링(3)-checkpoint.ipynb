# Instagram 크롤링 복습

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import re
```

```python
def insta_searching(word):
    url = "https://www.instagram.com/explore/tags/"+word
    return url
```

```python
def select_first(driver):
    first = driver.find_element_by_css_selector("._9AhH0")
    first.click()
    time.sleep(2)
# select_first(driver)
```

```python
#게시글 다양한 정보 
def get_content(driver):
    html = driver.page_source
    soup = BeautifulSoup(html, "lxml")
    
    # 게시글 읽기
    try:
        content = soup.select("div.C4VMK > span")[0].text
    except:
        content = "" 

    ## content에서 해시태그만 가져오기
    hashtag =re.findall("#[\w,가-힣ㄱ-ㅎ]+", content) 
    
    #날짜 가져오기
    date = soup.select("time._1o9PC.Nzb55")[0]['datetime'][:10]
    
    #좋아요 수 가져오기
    like = soup.select("div.Nm9Fw > button > span")[0].text
    
    #위치 정보 가져오기
    place = soup.select("div.M30cS")[0].text
        
    data = [content, date, like, place, hashtag]
    return data
```

```python
driver = webdriver.Chrome("C://mychrome/chromedriver.exe")
driver.get("https://www.instagram.com")
```

```python
word = "제주도맛집"
url = insta_searching(word)
driver.get(url) #한글을 유니코드 형식으로 읽어주는 함수 : get
time.sleep(2)
```

```python
select_first(driver)
```

# 스타벅스 데이터 분석

## - 크롤링

```python
#스타벅스 홈페이지(http://www.istarbucks.co.kr) -> 매장정보 -> 크롤링
#서울시 열린데이터 광장 -> 인구통계 데이터
#open api(application programming interface : 데이터 분석을 위해 데이터를
#제공할 수 있도록 만든 인터페이스)
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import re
import pandas as pd
```

```python
browser = webdriver.Chrome("C://mychrome/chromedriver.exe")
```

```python
url = "https://www.starbucks.co.kr/store/store_map.do?disp=locale"
browser.get(url)
```

```python
#서울시 버튼
seoulBtn = "div.loca_step1_cont > ul > li:nth-child(1) > a"
browser.find_element_by_css_selector(seoulBtn).click()
```

```python
#서울시 > 전체버튼
allBtn = "#mCSB_2_container > ul > li:nth-child(1) > a"
browser.find_element_by_css_selector(allBtn).click()
```

```python
html = browser.page_source
soup = BeautifulSoup(html, "html.parser")
```

```python
stores = soup.select("li.quickResultLstCon")

stores_li = []
for store in stores:
    name = store.select("strong")[0].text.strip()
    lat = store['data-lat'].strip()
    lng = store['data-long'].strip()
    store_type = store.select("i")[0]['class'][0][4:]
    
    addr_tel = str(store.select("p.result_details")[0]).split('<br/>')
    addr = addr_tel[0].split('>')[1]
    tel = addr_tel[1].split('<')[0]

    stores_li.append([name, lat, lng, store_type, addr, tel])
stores_li
```

## - 데이터 프레임

```python
columns = ['매장명', '위도', '경도', '매장타입', '주소', '전화번호']
df = pd.DataFrame(stores_li,
                  columns=columns,
                  index = list(i for i in range(1, len(stores_li)+1)))
```

```python
df.head() #상위 5개의 정보만 보여줌
```

```python
df.tail() #하위 5개의 정보만 보여줌
```

```python
df.info() #데이터 프레임의 정보를 출력
#null은 결측값, non-null은 누락값(결측값)이 없다는 뜻
#Dtype : 데이터타입 -> object 는 string과 같은 타입
```

## - 엑셀파일로 저장

```python
# 일반화 함수를 이용한 시군구 open api 호출
```

```python
def seoul_open_api_data(url, service):
    data_list = None
    try : 
        result_dict = requests.get(url).json()
        result_data = result_dict[service]
        code = result_data['RESULT']["CODE"]
        if code == 'INFO-000':
            data_list = result_data['row']
    except Exception as e:
        print(e)
    return data_list
```

```python
sgg_url = "http://openapi.seoul.go.kr:8088/{}/json/SdeTlSccoSigW/1/25/".format(SEOUL_API_AUTH_KEY)
sgg_data_list = seoul_open_api_data(sgg_url,'SdeTlSccoSigW')
if sgg_data_list == "None":
    print("키를 잘못 입력")
else:
    print(seoul_data_list)
```

```python
sggDf = pd.DataFrame(columns=["SIG_CD", 'SIG_KOR_NM', 'LAT', 'LNG'], data = sgg_data_list)
sggDf
```

```python
#dic에 없는 column의 이름을 쓰면 NaN값이 나옴 -> .columns로 세팅해줘야함 
#예 ) sggDf = pd.DataFrame(columns=["SIG_CD", 'SIG_KOR_NM', '위도', 'LNG'], data = sgg_data_list)
sggDf.columns = ["시군구코드", "시군구명", "위도", "경도"]
sggDf
```

## - 데이터 병합

```python
import re
```

```python
#스타벅스 매장 목록 엑셀 파일 불러오기
seoul_starbucks = pd.read_excel("seoul_starbucks_list.xlsx")
seoul_starbucks.head()
```

```python
#스타벅스 df에서 주소컬럼만 추출
sgg_names = []
for addr in seoul_starbucks['주소']:
    sgg = addr.split()[1]
    sgg_names.append(sgg)

seoul_starbucks['시군구명'] = sgg_names
seoul_starbucks.head()

seoul_starbucks.to_excel("seoul_starbucks_list.xlsx", index=False)
```

```python
#데이터 병합
```

```python
#스타벅스 데이터(매장명 | 위도 | 경도 | 매장타입 | 주소 | 전화번호)
seoul_starbucks = pd.read_excel("seoul_starbucks_list.xlsx")
seoul_starbucks
```

```python
#구 별로 스타벅스 매장 개수 세기 (피봇함수 이용)
"""
피봇(pivot)함수 : 피봇팅(특정 기준에 따른 데이터 요약)
############## 예 #############################
           스타벅스 매장 수
시군구명  |     
강남구    |      200
"""
starbucks_sgg_count = seoul_starbucks.pivot_table(index='시군구명', values="매장명", aggfunc='count').rename(columns={
    '매장명' : '스타벅스_매장수'
}) #aggfunc : 집계함수
```

```python
starbucks_sgg_count = starbucks_sgg_count.drop(['67'])
starbucks_sgg_count
```

```python
#시군구 목록 데이터 (시군구 코드 | 시군구명 | 위도 | 경도)
seoul_sgg = pd.read_excel('seoul_sgg_list.xlsx')
```

```python
# merge() : 두개의 데이터 프레임을 합치는 함수
# how는 어느쪽 join할 것인지. 왼쪽이 기준이 되어 합치기(merge)
seoul_sgg_final = pd.merge(seoul_sgg, starbucks_sgg_count, how='left', on='시군구명')
seoul_sgg_final
```

```python
#시군구별 인구통계 엑셀 불러오기
seoul_sgg_pop = pd.read_excel("sgg_pop.xlsx")
```

```python
seoul_sgg_final = pd.merge(seoul_sgg_final, seoul_sgg_pop, how='left', on='시군구명')
seoul_sgg_final
```

```python
# 서울시 사업체 데이터 엑셀파일 불러오기
seoul_sgg_biz = pd.read_excel("sgg_biz.xlsx")
```

```python
seoul_sgg_final = pd.merge(seoul_sgg_final, seoul_sgg_biz, how='left', on='시군구명')
```

```python
seoul_sgg_final.head()
```

```python
seoul_sgg_final.to_excel("seoul_sgg_stat.xlsx", index=False)
```

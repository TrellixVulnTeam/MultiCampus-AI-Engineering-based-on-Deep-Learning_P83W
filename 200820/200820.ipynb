```python
#각 열에 대해 (최대값 - 최소값) 결과를 추출
df['b'].max() - df['b'].min()
df['e'].max() - df['e'].min()
df['d'].max() - df['d'].min()
```

```python
#apply함수 : 각 컬럼에 해당되는 1차원 배열에 대해 함수를 적용
#df.apply(함수이름)

# f=lambda x:
# df.apply(f)
#df변수에 저장되어 있는 데이터 프레임에 대해 각각의 열을 f함수에 전달해라(적용해라)

f = lambda x : x.max()-x.min()
print(df.apply(f)) #axis=0 이거나 생략 : 데이터프레임의 행(row)을 따라 연산 수행
print(df.apply(f, axis=1)) #axis='columns'와 같은 표현
```

```python
#df에 저장되어 있는 데이터에 대해 도시 단위로 평균값을 구하시오.
f2 = lambda x : x.mean()
print(df.apply(f2, axis=1))
```

```python
#map함수 적용
srAge = pd.Series([10, 13, 15, 23, 22, 36])
srAge

#map함수를 이용하여 나이의 범위에 따라 10대, 20대, 30대로 변경하는 함수를 생성하시오
#0 10대
#1 10대
#2 10대
#3 20대
#4 20대
#5 30대

def myFunc(age):
    if age>=10 and age<20:
        sra = '10대'
    elif age>=20 and age<30:
        sra = '20대'
    elif age>=30 and age<40:
        sra = '30대'
    else:
        sra = '40대'
    return sra

print(srAge.map(myFunc))
```

```python
#apply 함수
df = pd.DataFrame([[1,3,5], [4,5,6], [7,8,10]])
print(df)

def diff(df):
    return df.max()-df.min()

#행 단위로 최대값과 최소값의 차이를 구하는 함수
print(df.apply(diff, axis=0))

#열 단위로 최대값과 최소값의 차이를 구하는 함수
print(df.apply(diff, axis=1))
```

```python
"""
         0        1        2
최대값    7        8       10
최소값    1        3        5
차이      6        5        5
"""
def diff2(df):
    return pd.Series([df.max(), df.min(), df.max()-df.min()], index=['최대값', '최소값','차이'])
```

```python
df.apply(diff2)
```

```python
df.apply(diff2, axis=1)
```

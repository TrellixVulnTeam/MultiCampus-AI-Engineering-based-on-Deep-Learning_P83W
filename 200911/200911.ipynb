```python
tesla = pd.read_csv("../etc/data/tesla_stock_quandl.csv")
#카페 -> 3041번 게시글 -> 다운로드
print(tesla.info())
tesla.head()
```

```python
tesla['Date'] = pd.to_datetime(tesla['Date'])
```

```python
tesla.info()
```

```python
#Date열에서 2010년만 출력
tesla['Date'][(tesla['Date'].dt.year==2010)&(tesla['Date'].dt.month==6)]
#Date열에서 2010년 6월만 출력
tesla[(tesla['Date'].dt.year==2010)&(tesla['Date'].dt.month==6)]
```

```python
#date컬럼 값을 행 인덱스로 설정하시오
# set_index : 기존의 행 인덱스를 제거하고 데이터 열 중 하나를 인덱스로 설정
# reset_index : 기존의 행 인덱스를 제거하고 인덱스를 데이터 열로 추가

#tesla.index = tesla['Date']
tesla = tesla.set_index("Date")
tesla.head()
```

```python
# tesla['refDate'] = 테슬라 데이터의 최초일부터 경과된 날짜 저장
tesla['refDate'] = tesla.index - tesla.index.min()
tesla
```

```python
#거래량(Volume)이 가장 많았던 날짜를 출력하시오.
print(tesla[tesla['Volume']==tesla['Volume'].max()].index.values)
```

# 1. 숫자 이미지 찾기

```python
import numpy as np
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
```

digits.data[0] ~ digits.data[9] 에는 숫자 0~9까지 데이터 저장  


digits.data[0] ~ [9]까지 각각의 데이터와

최대 내적을 갖는 숫자이미지 데이터를 화면에 출력(가장 유사한 이미지)

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
target = digits.target
```

```python
samples = np.arange(1797)
vec1 = []
vec2 = []
for i in range(len(samples)):
    #1~9까지 벡터 저장
    if i < 10:
        vec1.append(digits.images[samples[i]].reshape(64, 1))
    #비교할 나머지 사진들 벡터 저장
    else:
        vec2.append(digits.images[samples[i]].reshape(64, 1))
```

```python
gs = gridspec.GridSpec(1,2)
```

```python
for j, v1 in enumerate(vec1):
    temp = []
    for i, v2 in enumerate(vec2):
#         print((v1.T @ v2)[0][0])
        temp.append([i, (v1.T @ v2)[0][0]])
    
    stemp = sorted(temp, key=lambda x:x[1], reverse=True)
#     print(stemp)
    maxidx = stemp[0][0] #가장 근접한 이미지의 인덱스 찾기
#     print(vec2[maxidx])
    
    
    plt.figure(figsize=(8,4))
#     print("가장 근접한 인덱스 : ", maxidx+10)
#     plt.suptitle("Most Similar with {}th Image : Index {}th Image".format(j, maxidx+10))
#     plt.subplot(gs[0])
#     plt.title("[index : {}]".format(j))
#     plt.imshow(v1.reshape(8,8),cmap=plt.cm.bone_r)
#     plt.subplot(gs[1])
#     plt.title("[index : {}]".format(maxidx+10))
#     plt.imshow(vec2[maxidx].reshape(8,8),cmap=plt.cm.bone_r)
    fig, axs = plt.subplots(1, 2)
    fig.suptitle("Most Similar with {}th Image : Index {}th Image".format(j, maxidx+10))
    
    axs[0].set_title("[index : {}]".format(j))
    axs[0].set_xlabel("[target : {}]".format(target[j]))
    axs[0].imshow(v1.reshape(8,8),cmap=plt.cm.bone_r)
 
    axs[1].set_title("[index : {}]".format(maxidx+10))
    axs[1].set_xlabel("[target : {}]".format(target[maxidx+10]))
    axs[1].imshow(vec2[maxidx].reshape(8,8),cmap=plt.cm.bone_r)
    
```

# 2. 얼굴 이미지 분류 비교

```python
from sklearn.datasets import fetch_olivetti_faces
```

```python
data = fetch_olivetti_faces()
target = data.target
```

```python
samples = data.images[100].reshape(4096, 1)
vec = []
for i in range(400):
    vec.append(data.images[i].reshape(4096, 1))
```

```python
temp = []
for i, v in enumerate(vec):
    if i != 100:
        temp.append([i, (samples.T @ v)[0][0]])
        
stemp = sorted(temp, key=lambda x:x[1], reverse=True)
idx = stemp[0][0] #가장 근접한 이미지의 인덱스 찾기  

plt.figure(figsize=(8,4))

fig, axs = plt.subplots(1, 2)
fig.suptitle("Most Similar Image with sample")

axs[0].set_title("[Sample Image]")
axs[0].imshow(samples.reshape(64,64),cmap=plt.cm.bone)

axs[1].set_title("[index : {}]".format(idx))
axs[1].imshow(vec[idx].reshape(64,64),cmap=plt.cm.bone)
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("train.csv")
```

```python
df.info()
```

## 1. Name 컬럼에서 호칭 추출(Mr., Mrs. , ... Dr. , Sir, ...) using 정규표현식  
    
    ex)Mr : 0, Mrs : 1, 나머지 : 2로 변경하여, gender 컬럼을 추가하여 값으로 저장

```python
import re
```

```python
df['gender'] = np.where(df.Name.str.match("[\w\s\,]+Mr.[\w\s]+"), 0, np.where(df.Name.str.match("[\w\s\,]+Mrs.[\w\s]+"), 1, 2))
```

```python
df
```

## 2. Age열에 대해 결측값은 해당 데이터의 성별을 토대로 전체 데이터에 대한 성별 값의 평균으로 대체

    ex)성별이 남성, age가 결측값인 경우에는 남성 age 평균값으로 대체...

```python
male_age_mean = df[df.Sex == 'male']['Age'].mean()
female_age_mean = df[df.Sex == 'female']['Age'].mean()
```

```python
df['Age'] = np.where((df.Sex=='male') & (df.Age.isnull()), male_age_mean, np.where((df.Sex=='female') & (df.Age.isnull()), female_age_mean, df['Age']))
```

```python
df
```

## 3. FamilySize 컬럼 추가, 컬럼값은 SibSp+Parch값으로 함

```python
df['FamilySize'] = df['SibSp']+df['Parch']
```

```python
df
```

## 4. Fare열에 대한 평균, 중위수를 구함(극단치 존재 여부 확인)

```python
print("Fare의 평균 :", df.Fare.mean())
print("Fare의 중위수 :", df.Fare.median())
```

## 5. Age열에 대해 중복된 데이터는 모두 제거(저장하지 마세요)

```python
df.drop_duplicates(['Age'], keep=False, inplace=False)
```

## 6. Cabin열에 대해 unique한 데이터를 추출

```python
df['Cabin'].unique()
```

## 7. Age열에 대해 value_counts함수 or cut함수를 이용하여 구간을 나누어 데이터 개수를 세기 
   -> 연령대별 생존율/사망율 구하기  
    ex) 0-19,20-39,40-9,60- 연속형 데이터 -> 카테고리형(범주) 데이터

```python
age_range = pd.cut(df['Age'], bins=[0,20,40,60,100])
age_range
```

```python
pd.value_counts(age_range)
```

## 8. Embarked 컬럼값에 대해 항구별 승선인원 세기(비율)

```python
df_embraked = pd.DataFrame(df['Embarked'].value_counts())
df_embraked['percentage'] = df['Embarked'].value_counts()/len(df['Embarked'])*100
```

```python
df_embraked
```

## 9. Survived컬럼에 대해 생존(1)/사망(0) 인원수 및 비율

```python
df_survived = pd.DataFrame(df['Survived'].value_counts())
df_survived['percentage'] = df['Survived'].value_counts()/len(df['Survived'])*100
```

```python
df_survived
```

## 10. 성별에 따른 생존율/사망율 구하기

```python
survived_sex = df[['PassengerId', 'Survived', 'Sex']].groupby(by = ['Sex', 'Survived']).count()
survived_sex
```

### 10-1. 여자 승객 생존율/사망율

```python
female_survived = survived_sex.loc['female']
female_survived.columns = ['female passenger']
female_survived['percentage(%)'] = female_survived/female_survived.sum()*100
female_survived
```

### 10-2. 남자 승객 생존율/사망율

```python
male_survived = survived_sex.loc['male']
male_survived.columns = ['male passenger']
male_survived['percentage(%)'] = male_survived/male_survived.sum()*100
male_survived
```

```python
#데이터 정보 불러오는 코드를 함수로 지정
def get_prod_items(prod_items):
    prod_data = []
    for item in prod_items:
        title = item.select("p.prod_name > a")[0].text.strip()
        speclist = item.select("div.spec_list")[0].text.strip()
        price = item.select("p.price_sect > a > strong")[0].text.strip().replace(",","") 

        prod_data.append([title, speclist, price])
    
    return prod_data
```

## 데이터 전처리
* raw data -> 가공(전처리) 필요

```python
company_list = []
product_list = []
print(len(data['상품명']))
for title in data['상품명']:
    title_info = title.split(" ", 1) #split("구분자 문자", 분리하려는 수)
    company_name = title_info[0]
    company_list.append(company_name)
    product_name = title_info[1]
    product_list.append(product_name)
```

```python
#데이터 전처리를 위해서는 도메인 지식(domain knowledge)가 필요
# -> 연관성 있는 데이터만 추출하기 위해서
# -> 도메인 지식이 없다면 통계적 분석이 필요(ex. 상관분석 등)
spec_list = data['스펙목록'][0].split(" / ")
category = spec_list[0]

for spec in spec_list:
    if '사용시간' in spec:
        use_time_spec_value = spec.split(" ")[1].strip()
    elif '흡입력' in spec:
        suction_spec_value = spec.split(" ")[1].strip()
print(use_time_spec_value)
print(suction_spec_value)
```

```python
category_list = []
use_time_list = []
suction_list = []

for spec_data in data['스펙목록']:
    spec_data_list = spec_data.split(" / ")    
    category_list.append(spec_data_list[0])
    
    use_time = None
    suction = None
    for spec in spec_data_list:
        if '사용시간' in spec:
            use_time = spec.split(" ")[1].strip()
        elif '흡입력' in spec:
            suction = spec.split(" ")[1].strip()  
        
    use_time_list.append(use_time)
    suction_list.append(suction)
            
# print(category_list)
# print(use_time_list)
# print(suction_list)
```

```python
#시간을 분단위로 변환해서 저장
def convert_time_minute(time):
    try:
        if '시간' in time:
            adj_hour, adj_min = 0, 0 
            hour, miniute = time.split("시간")
            adj_hour = 60*int(hour[0])
            if '분' in time:
                adj_min =  int(miniute.split("분")[0])
            return adj_hour+adj_min
        else:
            return int(time.split("분")[0])
    except:
        return None

if __name__ == '__main__':    
    new_use_time_list = []
    for time in use_time_list:
        time_value = convert_time_minute(time)
        new_use_time_list.append(time_value)

    print(new_use_time_list)
```
